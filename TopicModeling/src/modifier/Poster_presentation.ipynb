```python
import pandas as pd

# Example DataFrame setup
Procon = 'E:/sp2023stock/TopicModeling/data/Kim_Data/PROPCON.csv'

df = pd.read_csv(Procon) 
```

```python
df
```

```python
# Ensure the 'Date' column is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Create 'Year-Quarter' column
df['Year-Quarter'] = df['Date'].dt.to_period('Q').astype(str)

# Group by 'Year-Quarter' and sum the values for each topic
grouped_df = df.groupby('Year-Quarter').sum()

# Sum all topic values into a single column for each 'Year-Quarter'
grouped_df['Total'] = grouped_df.sum(axis=1)

# Reset index to make 'Year-Quarter' a column again
grouped_df.reset_index(inplace=True)

# Select only 'Year-Quarter' and 'Total' for the final result
final_result = grouped_df[['Year-Quarter', 'Total']]

print(final_result)

```
